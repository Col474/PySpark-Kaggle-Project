```python
# get the details of the local Spark environment
#import findspark
#findspark.init()
from pyspark.sql import SparkSession
```

```python
#Create a Spark Session
spark = SparkSession.builder \
        .appName('ProgBigData-Spark Setup - wk2') \
        .master('local[*]') \
        .getOrCreate()
```

# 4 Create a Simple Dataframe

```python
# Test the setup
# Create a very simple Spart DataFrame, and display the DF
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
df = spark.createDataFrame(data, ["Name", "Age"])
df.show()
```
